```python
def ger(A):
    n=A.dimensions()[0]
    B=matrix(n,n,lambda i,j:A[i,j].abs())
    R=[sum(B.row(i))-B[i,i] for i in range(n)]
    E=A.eigenvalues()
    plt=point([(i.real(),i.imag()) for i in E],color="red",size=30)
    for i in range(n):
        plt=plt+disk((A[i,i],0),R[i],(0,2*pi),alpha=0.2 ,rgbcolor=(i+2,i,i-1))
    print("the greshgorin discs and eigenvalues are")
    show(plt)
    return
A=random_matrix(ZZ,5,5)
ger(A)
```
